```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

%matplotlib inline
```

```python
morts ={
    "One skydive": 8.0,
    "One mile walking": 1/17,
    "One mile by car": 1/250,
    "One mile by GA\n[120kts, 4e-7]": (1e6*4e-7)/(1.5078*120),
    "One mile by jet": 1/1000,
    "One mile by train": 1/6000,

}
morts_df = pd.DataFrame(morts, index=[0])
morts_df
```

```python
ax = sns.barplot(x=list(morts.keys()), y=list(morts.values()), )
ax.set_yscale('log')
ax.set_xlabel('Activity')
ax.set_title('Micromort for Transport Modes')
ax.set_ylabel('Micromorts (Millionth Chance of Death')
ax.set_xticklabels(ax.get_xticklabels(), rotation=25)
ax.get_figure().tight_layout()
ax.get_figure().savefig('morts.svg')
```
